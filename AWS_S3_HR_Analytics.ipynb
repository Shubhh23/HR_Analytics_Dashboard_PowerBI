```python
import boto3
import pandas as pd
```

```python
# Create S3 client
s3 = boto3.client("s3")

bucket_name = "hr-analytics-shubh"
file_name = "HRDataset_v14.csv"

# Download file from S3
s3.download_file(bucket_name, file_name, "HRDataset_v14.csv")

print("Dataset downloaded successfully!")
```

```python
df = pd.read_csv("HRDataset_v14.csv")

df.head()
```

```python
#Check shape
df.shape
```

```python
#Check missing values
df.isnull().sum()
```

```python
#Check duplicates
df.duplicated().sum()
```

```python
#Remove duplicates
df = df.drop_duplicates()
```

```python
#Save cleaned data
df.to_csv("HRDataset_Cleaned.csv", index=False)
```
